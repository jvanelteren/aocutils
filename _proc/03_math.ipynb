```
def power(a,b,M=None):
    # computes a**b. Actually python pow does this with optional third argument
    res = 1
    while(b):
        if b % 2 == 1: 
            res = (res * a) % M if M else res * a
            print('res',res)
        a *= a
        b //= 2
    return res
power(3,2)
```
